```python
# Make a new Jupyter notebook named big_o_notation.ipynb ## teacher said not to
import math
from scipy.special import factorial
# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible
# Curves to graph
#and label the curve "O(1)"


n = np.array(range(1,20))
y1 = np.array(0 * n + 1)
plt.plot(n,y1, c ='blue', label ="O(1)")



#  and label the curve "O(log n)"
n2 = np.array(range(1,20))
y2=np.array(np.log(n))
plt.plot(n2,y2, c ='green', label ="O(log n)")






#  and label the curve "O(n)"
n3 = np.array(range(1,20))
y3 = np.array(n3)
plt.plot(n3,y3,c ='yellow', label ="O(n)")

#  and label it "O(n log n)"
n4 = np.array(range(1,20))
y4=np.array(np.log(n4) * n)
plt.plot(n4,y4,c ='orange', label = "O(n log n)")




#  and label it "O(n^2)"
n5 = np.array(range(1,20))
y5=np.array(n5**2)
plt.plot(n5,y5,c ='black', label = "O(n^2)")



#  and label it "O(2^n)"
n6 = np.array(range(1,20))
y6=np.array(2**n6)
plt.plot(n6,y6, c ='brown', label = "O(2^n)")




#  and label it "O(n!)"
n7 = np.array(range(1,20))
y7= np.array(factorial(n7))

plt.plot(n7,y7, c ='purple', label = "O(n!)")





#  and label it "O(n^n)"
n8 = np.array(range(1,20))
y8 = np.array(n8 ** n8)
plt.plot(n8,y8, c ='red', label = "O(n^n)")

# plt.ylim(0, .4) #manually change the y axis
# plt.xlim(15,18)
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title("Big O Notation")
plt.ylim(0, 20)
plt.xlim(1,20)

plt.legend()
plt.show()
# Bonus Write the code necessary to write your name on a chart. Use box letters.

# Bonus: use curves for letters in your name that have curves in them.
```
