```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pydataset import data

# Use seaborn's load_dataset function to load the iris database to answer the following questions:
df = sns.load_dataset('iris')


# What does the distribution of petal lengths look like?
sns.histplot(df.petal_length, kde= True)


# Is there a relationship between petal length and petal width?
sns.relplot(x =df.petal_length, y = df.petal_width, data = df)
##positive correlation





```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? 
#For this, you'll visualize two numeric columns through the lense of a categorical column.
sns.relplot(x =df.sepal_length, y = df.sepal_width, data = df, col = df.species)

##you might be able to pick out the setosas but the versicolors and the virginicas might be harder
```

```python
# Which features would be best used to predict species?
sns.relplot(x =df.petal_length, y = df.petal_width, data = df, col = df.species)

##the pedal with and length would be better to predict species because it is easier
##to see the differences in species based off of the sizes
```
